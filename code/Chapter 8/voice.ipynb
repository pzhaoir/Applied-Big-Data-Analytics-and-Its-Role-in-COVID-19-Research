# Feature extraction

```python
# Extract Mel-frequency cepstral coefficients (MFCC), Spectral Centroid, Zero Crossing Rate, Chroma Frequencies, and Spectral Roll-off.

import csv

header = 'id chroma_stft rmse spectral_centroid spectral_bandwidth rolloff zero_crossing_rate'
for i in range(1, 21):
    header += f' mfcc{i}'
header = header.split()

file = open('features.csv', 'w', newline='')
writer = csv.writer(file)
writer.writerow(header)

for case_id in data["id"]:
    case_aud = "G:\\20210618\\20210618\\" + case_id + "\cough-heavy.wav"
    samples, sample_rate = librosa.load(case_aud)
    rmse = librosa.feature.rms(y=samples)
    chroma_stft = librosa.feature.chroma_stft(y=samples, sr=sample_rate)
    spec_cent = librosa.feature.spectral_centroid(y=samples, sr=sample_rate)
    spec_bw = librosa.feature.spectral_bandwidth(y=samples, sr=sample_rate)
    rolloff = librosa.feature.spectral_rolloff(y=samples, sr=sample_rate)
    zcr = librosa.feature.zero_crossing_rate(samples)
    mfcc = librosa.feature.mfcc(y=samples, sr=sample_rate)
    to_append = f'{case_id} {np.mean(chroma_stft)} {np.mean(rmse)} {np.mean(spec_cent)} {np.mean(spec_bw)} {np.mean(rolloff)} {np.mean(zcr)}'    
    for e in mfcc:
        to_append += f' {np.mean(e)}'
    file = open('features.csv', 'a', newline='')
    with file:
        writer = csv.writer(file)
        writer.writerow(to_append.split())
```

```python
features = pd.read_csv("features.csv")
features
```

```python
# add test_status lable 

class_label = data[["id","test_status"]]
feature_class = features.merge(class_label, on = "id", how = "inner")
feature_class.loc[(feature_class.test_status != 'p'),'test_status']='n'
feature_class
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split 

# convert features and corresponding classification labels into numpy arrays

df = feature_class.iloc[:,1:]
X = np.array(df.iloc[:, :-1])
y = np.array(df.test_status.tolist())

# encode the classification labels

le = LabelEncoder()
y = le.fit_transform(y)

# scale features

scaler = StandardScaler()
X = scaler.fit_transform(np.array(X, dtype = float))


# split the dataset 

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
list(y).count(1)
```

```python
from keras import layers
import keras
from keras.models import Sequential

model = Sequential()
model.add(layers.Dense(26, activation='relu', input_shape=(x_train.shape[1],)))
model.add(layers.Dense(18, activation='relu'))
model.add(layers.Dense(12, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
classifier = model.fit(x_train,
                    y_train,
                    epochs=100,
                    batch_size=128)
```

```python
from sklearn.metrics import accuracy_score
y_pred = model.predict_classes(x_test)
print("Accuracy {:0.2f}%.".format(accuracy_score(y_test, y_pred)*100))
```
