```python
# Load data for 2021.06.18

import pandas as pd

data = pd.read_csv("G:\20210618\20210618.csv")
data
```

```python
# positive case and negative case

pos_case = data.loc[data["test_status"]=="p"].reset_index()
neg_case = data.loc[data["test_status"]=="n"].reset_index()
pos_case
```

```python
neg_case
```

# Cough

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

pname = "G:\\20210618\\20210618\\" + pos_case["id"][0] + "\cough-heavy.wav"
nname = "G:\\20210618\\20210618\\" + neg_case["id"][0] + "\cough-heavy.wav"

print("Cough spectrogram for COVID-19 positive case:")
samples, sample_rate = librosa.load(pname)

fig = plt.figure(figsize=[10,7])
ax = fig.add_subplot(111)
ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)
ax.set_frame_on(False)
S = librosa.feature.melspectrogram(y=samples, sr=sample_rate)
librosa.display.specshow(librosa.power_to_db(S, ref=np.max))
plt.savefig(r"G:\pos-cough-spec.png", dpi=300)
```

```python
print("Cough spectrogram for COVID-19 negative case:")
samples, sample_rate = librosa.load(nname)

fig = plt.figure(figsize=[10,7])
ax = fig.add_subplot(111)
ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)
ax.set_frame_on(False)
S = librosa.feature.melspectrogram(y=samples, sr=sample_rate)
librosa.display.specshow(librosa.power_to_db(S, ref=np.max))
plt.savefig(r"G:\neg-cough-spec.png", dpi=300)
```
