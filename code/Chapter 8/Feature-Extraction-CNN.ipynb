```python
# Load data for 2021.06.18

import pandas as pd

data = pd.read_csv(r"G:\20210618\20210618.csv")
data
```

```python
data = data.loc[data["test_status"]!="na"]
data.shape
```

```python
cough_voice = ["cough-heavy.wav", "cough-shallow.wav"]
breath_voice = ["breathing-deep.wav", "breathing-shallow.wav"]
vowel_voice = ["vowel-a.wav", "vowel-e.wav", "vowel-o.wav"]
```

# Cough

```python
# Extract Mel-frequency cepstral coefficients (MFCC), Spectral Centroid, Zero Crossing Rate, Chroma Frequencies, and Spectral Roll-off.

import csv
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

header = 'id chroma_stft rmse spectral_centroid spectral_bandwidth rolloff zero_crossing_rate'
for i in range(1, 21):
    header += f' mfcc{i}'
header = header.split()

file = open(r'G:\voice-training\c-features.csv', 'w', newline='')
writer = csv.writer(file)
writer.writerow(header)

for case_id in data["id"]:
    for voice in cough_voice: 
        case_aud = "G:\\20210618\\20210618\\" + case_id + "\\"+ voice
        samples, sample_rate = librosa.load(case_aud)
        rmse = librosa.feature.rms(y=samples)
        chroma_stft = librosa.feature.chroma_stft(y=samples, sr=sample_rate)
        spec_cent = librosa.feature.spectral_centroid(y=samples, sr=sample_rate)
        spec_bw = librosa.feature.spectral_bandwidth(y=samples, sr=sample_rate)
        rolloff = librosa.feature.spectral_rolloff(y=samples, sr=sample_rate)
        zcr = librosa.feature.zero_crossing_rate(samples)
        mfcc = librosa.feature.mfcc(y=samples, sr=sample_rate)
        to_append = f'{case_id} {np.mean(chroma_stft)} {np.mean(rmse)} {np.mean(spec_cent)} {np.mean(spec_bw)} {np.mean(rolloff)} {np.mean(zcr)}'    
        for e in mfcc:
            to_append += f' {np.mean(e)}'
        file = open(r'G:\voice-training\c-features.csv', 'a', newline='')
        with file:
            writer = csv.writer(file)
            writer.writerow(to_append.split())
```

```python
# add test_status lable 

class_label = data[["id","test_status"]]
class_label = class_label.loc[class_label["test_status"]!="na"]
class_label.shape
```

```python
v_feature = pd.read_csv(r"G:\voice-training\v-features.csv")
c_feature = pd.read_csv(r"G:\voice-training\c-features.csv")
b_feature = pd.read_csv(r"G:\voice-training\b-features.csv")
```

```python
from functools import reduce

merge_data = pd.concat([v_feature, c_feature, b_feature], ignore_index=True)
merge_data.head()
```

```python
# add test_status label

merge_data = merge_data.merge(class_label, on = "id", how = "left")
merge_data
```

# CNN model

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split 

# convert features and corresponding classification labels into numpy arrays

df = merge_data.iloc[:,1:]
X = np.array(df.iloc[:, :-1])
y = np.array(df.test_status.tolist())

# encode the classification labels

le = LabelEncoder()
y = le.fit_transform(y)

# scale features

scaler = StandardScaler()
X = scaler.fit_transform(np.array(X, dtype = float))


# split the dataset 

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
x_train = x_train.reshape((x_train.shape[0], x_train.shape[1], 1))
x_test = x_test.reshape((x_test.shape[0], x_test.shape[1], 1))
print('Train set shape', x_train.shape)
print('Test set shape', x_test.shape)
```

```python
from keras import layers
import keras
from keras.models import Sequential
from keras.layers import Flatten, Dense
from keras.layers.convolutional import Conv1D, MaxPooling1D

# Initialising the CNN
model = Sequential()
model.add(Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(x_train.shape[1], x_train.shape[2])))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(50, activation='relu'))
model.add(Dense(1))
model.compile(loss='mse', optimizer="adam", metrics=['acc'])
model.summary()
```

```python
# Training the model
cnn_history = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=30, verbose=2)
```

```python
plt.figure(figsize=(8,8))
plt.title("Loss Value")
plt.plot(cnn_history.history['loss'])
plt.plot(cnn_history.history['val_loss'])
plt.legend(['loss', 'val_loss'])
print('loss:', cnn_history.history['loss'][-1])
print('val_loss:', cnn_history.history['val_loss'][-1])
```

```python
y_pred = model.predict_classes(x_test)
```

```python
import sklearn.metrics as metrics
import matplotlib.pyplot as plt

# calculate the fpr and tpr for all thresholds of the classification
probs = model.predict_proba(x_test)
preds = probs[:,0]
fpr, tpr, threshold = metrics.roc_curve(y_test, preds)
roc_auc = metrics.roc_auc_score(y_test, preds)

plt.figure(figsize=(8,6))
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
#plt.savefig(r"G:\roc-curve-cnn.png", dpi=300)
```

```python
from sklearn.metrics import accuracy_score

y_pred_class = model.predict_classes(x_test)
print("Accuracy = {0:.2f}%".format(accuracy_score(y_test, y_pred_class)*100))
```
