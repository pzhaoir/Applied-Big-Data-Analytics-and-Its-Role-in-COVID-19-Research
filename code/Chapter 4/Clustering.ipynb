```python
merge_clu = pd.merge(google_us_c, us_case, how="left", on=["state","county","date"])
merge_clu
```

```python
merge_clu.loc[merge_clu["state"]=="Wyoming"]
```

```python
import math 

state_list = []
county_list = []
start_date = []
end_date = []
mean_cases = []
mean_parks = []

for state in list(set(merge_clu["state"])):
    df_state = merge_clu.loc[merge_clu["state"]==state]
    for county in list(set(df_state["county"])):
        df_state_county = df_state.loc[df_state["county"]==county]
        start = min(df_state_county["date"])
        start_date.append(start)
        end = max(df_state_county["date"])
        end_date.append(end)
        mean_p = df_state_county["parks"].mean()
        mean_parks.append(mean_p)
        mean_c = df_state_county["cases"].mean()
        mean_cases.append(mean_c)
        state_list.append(state)
        county_list.append(county)
    
df_mean = pd.DataFrame(data={"state": state_list, "county":county_list,"start_date": start_date, "end_date": end_date,"mean_parks":mean_parks,
                            "mean_cases":mean_cases})
df_mean
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler


features = merge_mobility_ier[["COVID-19_mobility_pool_indicator","IER"]]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
# chose the approporiate cluster number

kmeans_kwargs = {
    "init":"random",
    "n_init":15,
    "max_iter":300,
    "random_state":42
}

# A list holds the SSE values for each k
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(scaled_features)
    sse.append(kmeans.inertia_)
    
plt.figure(figsize=(7,5))
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
#plt.savefig("kmean-curve.png")
#plt.savefig("kmean-curve.tiff")
```

```python
kmeans = KMeans(
    init="random",
    n_clusters=3,
    n_init=10,
    max_iter=300,
    random_state=42
    )

labels = kmeans.fit_predict(scaled_features)
plt.figure(figsize=(30,20))
plt.scatter(scaled_features[:, 0], scaled_features[:, 1], c=labels,
            s=50, cmap='viridis')
for i in range(len(list(merge_mobility_ier["state"]))):
    plt.text(scaled_features[:,0][i],scaled_features[:,1][i],"{}".format(list(merge_mobility_ier["state"])[i]),size=17)
    
plt.xlabel("COVID-19 mobility pool indicator",fontsize=17)
plt.ylabel("IER",fontsize=17)
    
#plt.savefig("kmean-clustering-2.png")
#plt.savefig("kmean-clustering-2.tiff")
```

```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

#data, labels = load_digits(return_X_y=True)
#(n_samples, n_features), n_digits = data.shape, np.unique(labels).size


reduced_data = PCA(n_components=2).fit_transform(scaled_features)
kmeans = KMeans(init="k-means++", n_clusters=3, n_init=4)
kmeans.fit(reduced_data)

# Step size of the mesh. Decrease to increase the quality of the VQ.
h = .02     # point in the mesh [x_min, x_max]x[y_min, y_max].

# Plot the decision boundary. For that, we will assign a color to each
x_min, x_max = reduced_data[:, 0].min() - 1, reduced_data[:, 0].max() + 1
y_min, y_max = reduced_data[:, 1].min() - 1, reduced_data[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Obtain labels for each point in mesh. Use last trained model.
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(1)
plt.clf()
plt.imshow(Z, interpolation="nearest",
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap=plt.cm.Paired, aspect="auto", origin="lower")

plt.plot(reduced_data[:, 0], reduced_data[:, 1], 'k.', markersize=2)
# Plot the centroids as a white X
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], marker="x", s=169, linewidths=3,
            color="w", zorder=10)
#plt.title("K-means clustering on the digits dataset (PCA-reduced data)\n"
          #"Centroids are marked with white cross")
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())

#plt.savefig("kmean-boundary.png")
#plt.savefig("kmean-boundary.tiff")
```
