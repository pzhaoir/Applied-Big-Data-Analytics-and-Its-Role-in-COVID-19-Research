# Waze Report

```python
import pandas as pd

waze_data = pd.read_csv("waze_mobility.csv")
waze_data
```

```python
list(set(waze_data["country"]))
```

```python
waze_us = waze_data.loc[(waze_data["country"]=="United States")&(waze_data["city"]!="Total")]
waze_us.head()
```

```python
df_at = waze_us.loc[waze_us["city"]=="Atlanta"]
df_at
```

```python
import math 

city_list = list(set(waze_us["city"]))
start_date = []
end_date = []
max_value = []
min_value = []
mean_value = []
std_value = []
vol_value = []

for city in city_list:
    df_city = waze_us.loc[waze_us["city"]==city]
    start = min(df_city["date"])
    start_date.append(start)
    end = max(df_city["date"])
    end_date.append(end)
    max_dr = max(df_city["driving_waze"])
    max_value.append(max_dr)
    min_dr = min(df_city["driving_waze"])
    min_value.append(min_dr)
    mean_dr = df_city["driving_waze"].mean()
    mean_value.append(mean_dr)
    std_dr = df_city["driving_waze"].std()
    std_value.append(std_dr)
    vol_dr = std_dr * math.sqrt(df_city.shape[0])
    vol_value.append(vol_dr)
    
df_vol = pd.DataFrame(data={"City": city_list, "Start_Date": start_date, "End_Date": end_date,"Max":max_value,"Min":min_value,"Mean":mean_value,
                            "Std": std_value, "volatility":vol_value})
df_vol
```

```python
df_vol.to_csv("data-kmean-training.csv")
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler


features = df_vol[["Mean","volatility"]]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
# chose the approporiate cluster number

kmeans_kwargs = {
    "init":"random",
    "n_init":15,
    "max_iter":300,
    "random_state":42
}

# A list holds the SSE values for each k
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(scaled_features)
    sse.append(kmeans.inertia_)
    
plt.figure(figsize=(7,5))
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
#plt.savefig("kmean-curve.png")
#plt.savefig("kmean-curve.tiff")
```

```python
kmeans = KMeans(
    init="random",
    n_clusters=3,
    n_init=10,
    max_iter=300,
    random_state=42
    )

labels = kmeans.fit_predict(scaled_features)
plt.figure(figsize=(25,17))
plt.scatter(scaled_features[:, 0], scaled_features[:, 1], c=labels,
            s=50, cmap='viridis')
for i in range(len(city_list)):
    plt.text(scaled_features[:,0][i],scaled_features[:,1][i],"{}".format(city_list[i]),size=15)
    
#plt.savefig("kmean-clustering.png")
#plt.savefig("kmean-clustering.tiff")
```

```python
import pickle
filename = 'city-kmeans_model.sav'
pickle.dump(kmeans, open(filename, 'wb'))
```

```python
import seaborn as sns

df_vol["label"]=labels
df_vol = df_vol.rename(columns={"volatility":"Volatility"})
plt.style.use("fivethirtyeight")
plt.figure(figsize=(15, 13))
scat = sns.scatterplot(
    x="Mean",
    y="Volatility",
    s=110,
    data=df_vol,
    hue="label",
    palette="Set2", style="label", legend= False
    )


for i in range(len(list(df_vol["City"]))):
    plt.text(df_vol["Mean"][i],df_vol["Volatility"][i],"{}".format(list(df_vol["City"])[i]),size=13)

#plt.savefig("kmean-city-clustering.png")
#plt.savefig("kmean-city-clustering.tiff")
```
